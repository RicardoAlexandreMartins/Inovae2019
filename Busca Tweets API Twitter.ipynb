```python
session = OAuth1Session(CHAVE_CONSUMO, TOKEN_CONSUMO,TOKEN_ACESSO, CHAVE_ACESSO)
```

```python
# Vamos criar uma fun√ß√£o para busca dos tweets
def BuscaTweets(query):
    url = url_busca + "?q=%s&lang=pt&count=100"
    url = url % (requests.utils.quote(query))
    response = session.get(url)
    return response.content
```

```python
# Vamos buscar os tweets primeiros em portugues 100 tweets que falam sobre um assunto
resposta = BuscaTweets('palmeiras nao tem mundial')
```

```python
tweets = json.loads(resposta)
```

```python
# Vamos criar agora uma fun√ß√£o para buscar os perfis
def BuscaPerfil(perfil):
    url = url_perfil + "?screen_name=" + perfil
    response = session.get(url)
    return response.content
```
